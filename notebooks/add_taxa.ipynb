```python
mappings = pd.read_parquet("~/datasets/ncbi_tax_mapping.parquet")
```

```python
mappings
```

```python
mappings.set_index("source", inplace=True)
```

```python
def get_mapping(id):
	try:
		m =  mappings.loc[id]
		return m["target"]
	except Exception:
		return id
	
ncbi_taxa = df["organism_identifier"].apply(get_mapping)
```

```python
classes = ncbi_taxa.apply(lambda x: taxa.loc[x]["class"])
```

```python
phylum =  ncbi_taxa.apply(lambda x: taxa.loc[x]["phylum"])
```

```python
df["ncbi_taxonomy_id"] = ncbi_taxa
```

```python
df["ncbi_taxonomy_class"] = classes
```

```python
df["ncbi_taxonomy_phylum"] = phylum
```
