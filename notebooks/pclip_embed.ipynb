```python
import pandas as pd
from protein_clip import load_proteinclip
```

```python
df = pd.read_parquet("../data/swissprot-embeddings-v2.parquet")
df
```

```python
import numpy as np
```

```python
m = load_proteinclip(6)
def esm_to_pclip(x):
    x_norm = x / np.linalg.norm(x)
    pc_embed = m.predict(x_norm)
    return pc_embed

df["proteinclip_embed_esm2_6"] = df["embedding_esm2_t6_8M_UR50D"].apply(esm_to_pclip)
```

```python
df = df.rename(columns={"esm2_embed_6": "embed_esm2_6"})
```

```python
df.to_parquet("../data/swissprot-embeddings-v3.parquet")
```
